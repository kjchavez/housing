```python
import geopandas as gpd
import geoplot as gplt
import matplotlib.pyplot as plt
import pandas as pd; pd.set_option('max_columns', 6)

%matplotlib inline
```

```python
zipcode_shp_file = "../data/raw/ZIPCODES_NT_POLY.shp"
zipcodes = gpd.read_file(zipcode_shp_file).to_crs({'init': 'epsg:4326'})
zipcodes.head()
```

```python
import geoplot.crs as gcrs
proj = gcrs.AlbersEqualArea()
central_mass = zipcodes.query('"01500" >= POSTCODE >= "01400"')
gplt.polyplot(central_mass, facecolor='lightgray', projection=proj)
```
